```python
page = requests.get('https://brawlify.com/ru/brawlers')
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
world_img = 'https://supercell.com/images/1a5b69311180a4a1c374e10556941f05/hero_bg_brawlstars.a385872a.webp'
world_name = 'brawl stars'
global_char_index = 1666
global_world_index = 25
```

```python
brawlers = []
i = 0
for brawler in soup.find_all(class_='d-flex flex-row align-items-center col-6 col-sm-4 col-md-3 col-lg-3 mb-2'):
    link = brawler.find('a');
    brawlers.append(
        {
            'name': link.attrs['title'].split(' is a')[0],
            'rarity':link.attrs['title'].split(' is a')[1].split()[0], 
            'href': 'https://brawlify.com' + link.attrs['href'],
            'image': brawler.find('img').attrs['data-src'],
            'index': i
        }
    )
    i = i + 1
```

```python
brawlers
```

```python
with open("brawlers.csv", 'w') as file:
    writer = csv.writer(file)
    for row in brawlers:
        writer.writerow([global_char_index+1+row['index'], row['index'], global_world_index+1, world_name, world_img, row['name'], row['image'], 'Игры', None, None, None])
```

```python
with open("data.csv", 'a') as file:
    writer = csv.writer(file)
    for row in brawlers:
        writer.writerow([global_char_index+1+row['index'], row['index'], global_world_index+1, world_name, world_img, row['name'], row['image'], 'Игры', None, None, None])
```
